```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Exploration

#### Age

```python
plt.hist(train.age)
plt.show()
```

```python
#is there any age difference between those whose outcome was a yes and a no?
plt.hist(train['age'][train.y == 'no'], label='no')
plt.hist(train['age'][train.y == 'yes'], label='yes')
plt.legend(loc='upper right')
plt.show()
```

NOTES:

- Can be binned into age brackets

### Balance

```python
plt.hist(train.balance)
plt.show()
```

```python
#is there any balance difference between those whose outcome was a yes and a no?
plt.hist(train['balance'][train.y == 'no'], label='no')
plt.hist(train['balance'][train.y == 'yes'], label='yes')
plt.legend(loc='upper right')
plt.show()
```

NOTES:
- Balances are highly right skewed. Will logging work to normalise them a bit?

```python
plt.hist(np.log10(train['balance'][train.balance > 0]))
plt.show()
```

```python
#balances less than 0?
print("There are", len(train.balance[train.balance < 0])," rows with negative balances out of", len(train.balance))
```

NOTES:
- Should we keep these negative balances? What is the distribution of the y variable of people with negative balances versus those with positives?

```python
train.y[train.balance < 0].value_counts(normalize = True)
```

```python
train.y[train.balance > 0].value_counts(normalize = True)
```

NOTES:
- At face value, the distributions look different so just deleting the negative balances might not help.
